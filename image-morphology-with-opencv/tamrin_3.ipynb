# Qestion 1: DILATION 

## part 1

### Step 1: Read the input binary image

```python
# The image is read in grayscale mode.
image_1 = cv2.imread(r"imgs\circle-square.png", cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image_1 is None:
    raise IOError(r"Error: Could not read the image 'imgs\circle-square.png'.")
```

### Step 2: Ensure the image is binary


```python
# Morphological dilation assumes a binary image where:
# - Foreground pixels are white (255)
# - Background pixels are black (0)
# We apply a threshold to guarantee a clean binary image.
_, binary_image_1 = cv2.threshold(image_1, 127, 255, cv2.THRESH_BINARY)
```

```python
binary_image_1
```

### Step 3: Define kernel sizes for comparison

```python
# Different kernel sizes allow us to observe how the amount of dilation increases with the scale of the structuring element.
kernel_sizes = [4, 7, 10, 15, 20]
```

### Step 4,5: Apply morphological dilation for each kernel size and save the result


```python
# Morphological Dilation:
# - Expands the white (foreground) regions in a binary image
# - Adds pixels to object boundaries
# The shape of the structuring element affects *how* objects grow,
# while the size affects *how much* they grow.
for size in kernel_sizes:
    kernel_dim = (size, size)

    # Square structuring element
    square_kernel = cv2.getStructuringElement(
        cv2.MORPH_RECT,
        kernel_dim
    )

    # Circular (disk-shaped) structuring element
    circular_kernel = cv2.getStructuringElement(
        cv2.MORPH_ELLIPSE,
        kernel_dim
    )

    # Apply dilation
    dilated_square = cv2.dilate(binary_image_1, square_kernel, iterations=1)
    dilated_circular = cv2.dilate(binary_image_1, circular_kernel, iterations=1)

    
    # Step 5: Save the results
    
    cv2.imwrite(f"square-circle-square-{size}.png", dilated_square)
    cv2.imwrite(f"square-circle-circle-{size}.png", dilated_circular)
```

## Part 2: Pepper noise removal

### Step 1: Read the input grayscale image


```python
# The image is read in grayscale mode . pepper noise typically appears as dark pixels in intensity images.
image_2 = cv2.imread(r"imgs\Cameraman.png", cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image_2 is None:
    raise IOError("Error: Could not read the image 'cameraman.png'.")
```

```python
image_2
```

### Step 2: Define multiple kernel sizes

```python
kernel_sizes = [3, 5, 7, 9, 11]
```

### Step 3,4 : Apply dilation using different kernels and save the results

```python
for size in kernel_sizes:
    kernel_dim = (size, size)

    # Circular (disk-shaped) structuring element
    #
    # Rationale:
    # - Circular kernels expand pixels uniformly in all directions
    # - They avoid directional artifacts that square kernels may introduce
    structuring_element = cv2.getStructuringElement(
        cv2.MORPH_ELLIPSE,
        kernel_dim
    )

    # Apply morphological dilation
    denoised_image = cv2.dilate(
        image_2,
        structuring_element,
        iterations=1
    )

   
    # Step 4 : Save the result for this kernel size
    
    cv2.imwrite(f"cameraman-denoised-{size}.png", denoised_image)
```

## Part 3: Edge detection

### Step 1: Read the input grayscale image


```python
image_3 = cv2.imread(r"imgs\Lady.png", cv2.IMREAD_GRAYSCALE)

if image_3 is None:
    raise IOError("Error: Could not read the image 'lady.png'.")
```

```python
image_3
```

### Step 2: Define multiple kernel sizes


```python
# Using multiple kernel sizes allows us to observe:
# - Small kernels  -> thin, fine edges
# - Large kernels  -> thicker, more prominent edges
kernel_sizes = [3, 5, 7, 9, 11]
```

### Step 3,4: Apply dilation and edge extraction for each kernel and save


```python
for size in kernel_sizes:
    kernel_dim = (size, size)

    
    # 1. Square structuring element (RECT)
   
    square_kernel = cv2.getStructuringElement(
        cv2.MORPH_RECT,
        kernel_dim
    )

    # Apply dilation with square kernel
    dilated_square = cv2.dilate(
        image_3,
        square_kernel,
        iterations=1
    )

    # Edge detection using morphological gradient idea
    edge_square = cv2.subtract(dilated_square, image_3)

    # Save result
    cv2.imwrite(f"lady-edge-square-{size}.png", edge_square)

    
    # 2. Circular structuring element (ELLIPSE)
    
    circular_kernel = cv2.getStructuringElement(
        cv2.MORPH_ELLIPSE,
        kernel_dim
    )

    # Apply dilation with circular kernel
    dilated_circular = cv2.dilate(
        image_3,
        circular_kernel,
        iterations=1
    )

    # Edge detection using morphological gradient idea
    edge_circular = cv2.subtract(dilated_circular, image_3)

    # Save result
    cv2.imwrite(f"lady-edge-circular-{size}.png", edge_circular)
```

# Question 2 : EROSION

## Part 1

### Step 1: Define multiple kernel sizes

```python
# Larger kernels result in stronger erosion effects.
kernel_sizes = [4, 7, 10, 15, 20]
```

### Step 2,3: Apply erosion using different kernels and saving


```python
for size in kernel_sizes:
    kernel_dim = (size, size)

    # Square structuring element
    square_kernel = cv2.getStructuringElement(
        cv2.MORPH_RECT,
        kernel_dim
    )

    # Circular (disk-shaped) structuring element
    circular_kernel = cv2.getStructuringElement(
        cv2.MORPH_ELLIPSE,
        kernel_dim
    )

    # Apply erosion
    eroded_square = cv2.erode(binary_image_1, square_kernel, iterations=1)
    eroded_circular = cv2.erode(binary_image_1, circular_kernel, iterations=1)

    # Step 3: Save results separately
    cv2.imwrite(
        f"circle-square-erode-square-{size}.png",
        eroded_square
    )

    cv2.imwrite(
        f"circle-square-erode-circle-{size}.png",
        eroded_circular
    )
```

## Part2: Salt and pepper noise removal

### Step 1: Define multiple kernel sizes


```python
kernel_sizes = [3, 5, 7, 9, 11]
```

### Step 2, 3: Apply erosion using different kernels and save the results


```python
for size in kernel_sizes:
    kernel_dim = (size, size)

    # Structuring element choice:
    # - Circular kernel avoids directional artifacts
    # - Small-to-medium sizes limit detail loss
    structuring_element = cv2.getStructuringElement(
        cv2.MORPH_ELLIPSE,
        kernel_dim
    )

    # Apply morphological erosion
    eroded_image = cv2.erode(
        image_2,
        structuring_element,
        iterations=1
    )

    
    # Step 3: Save the result for this kernel size
   
    cv2.imwrite(f"cameraman-erode-{size}.png", eroded_image)
```
